## Introduction

In biological systems, it is useful to model the rate at which reactions take place in order to undestand the kinetics of the system.  This can be accomplished by utilizing the law of mass action to model a set of ODE's, which are then used to solve the kinematics of the system.  By doing this a basic understanding of the system and the mechanics driving the interactions between the different sub-systems can be observed. In this experiment this was done on 3 different systems.  

The first was modeling a basic enzyme catalyzation reaction where a substrated is used to form a product through catalization with an enzyme.  this was done using the Michaelis-Menten equation.  This basic equation was then modified to include a inhibiting substrate which can reduce the rate of the reation.  Three different kinds of inhibition were examined: competitive, non-competitive and uncompetitive, and the reaction velocity (rate) was examined for each with various reactions velocities for the inhibitors, concentration of inhibitors and "saturation's" of inhibitors.  

In the second part of the experiment, a method for determining the the parameters in the enzyme kinimatics previously used was used on provided data to find the parameters for the Michaelis-Menten equation.  Initially the parameters for the Michaelis-Menten equation were estimated from examining the data after plotting, these were unclear due to the resolution required to find the parameters.  As such another method was used, the Lineweaver - Burk plot, which plots $1/V$ against $1/S$ to find the parameters.  The error against the data was examined for both of these methods.  

In the final part of experiment, ionic channels in a celular membrain were examined as if they were enzymes for mediating ion transport over a membrane.  The system was sketched and modeled in order to solve the ODE's for the system.  This was then varried for different initial conditions and different parameter values.  

## 1 Michaelis-Menten Kinetics

The law of mass action, which was introduced by Guldberg and Wagge and states that the reaction rate is proportional to the probability of a collision occuring between the reactants.  

```python
# Units:
# k_m is how efficient the enzyme/substrate affinity,  micro moles
# V_max is the speed at which the reaction goes, micro moles/s

def michaelis_meten_kinetics_eq(V_max, S, k_m):
    V = []
    for s in S:
        V.append((V_max*s)/(s+k_m))
    return V
```

```python
# Setting up the variables

x = np.arange(0, 200, 0.5)
K_list = np.arange(10, 100, 5)
V_list = np.arange(25, 100, 5)

y_vark = []
y_varv = []

plt.figure(0, figsize = (10.0, 8.0))
plt.title('Michaelis-Menten Kinetics with Various Max Velocity')
plt.xlabel('Substrate Concentration ($\mu M$)')
plt.ylabel('Reaction Max Speed ($\mu M/s$)')

for v in V_list:
    tmp = michaelis_meten_kinetics_eq(v, x, 10)
    y_varv.append(tmp)
    plt.plot(x,tmp,label = "$V_{max} = $"+str(v))
plt.legend(ncol=5)
```

```python
plt.figure(1, figsize = (10.0, 8.0))
plt.title('Michaelis-Menten Kinetics with Various Michaelis Constant')
plt.xlabel('Substrate Concentration ($\mu M$)')
plt.ylabel('Reaction Max Speed ($\mu M/s$)')

for k in K_list:
    tmp = michaelis_meten_kinetics_eq(25, x, k)
    y_vark.append(tmp)
    plt.plot(x,tmp,label = "$K_{m} = $"+str(v))
plt.legend(ncol=3)
```

```python
# Inhibition Kinetics
# Units:
# k_m is how efficient the enzyme/substrate affinity,  micro moles
# V_max is the speed at which the reaction goes, micro moles/s

def competitve_kinetics_eq(I, S, ki):
    V_max = 12
    k_m = 3
    V = []
    for s in S:
        V.append((V_max*s)/(s+(k_m*(1+(I/ki)))))
    return V

def non_competitve_kinetics_eq(I, S, ki):
    V_max = 12
    k_m = 3
    V = []
    for s in S:
        V.append((V_max*s)/((1+(I/ki))*(s+k_m)))
    return V
                 
def un_competitve_kinetics_eq(I, S, ki):
    V_max = 12
    k_m = 3
    V = []
    for s in S:
        V.append((V_max*s)/(s*(1+(I/ki))+k_m))
    return V
```

```python
I_list = np.arange(2, 22, 2)
Ki_list = np.arange(1, 10, .5)

y_varI_c = []
y_vark_c = []

y_varI_nc = []
y_vark_nc = []

y_varI_uc = []
y_vark_uc = []

plt.figure(0, figsize = (10.0, 8.0))
plt.title('Michaelis-Menten Competitive Kinetics with Various Inhibitor Staturation Constants')
plt.xlabel('Substrate Concentration ($\mu M$)')
plt.ylabel('Reaction Max Speed ($\mu M/s$)')

for k_i in Ki_list:
    tmp = competitve_kinetics_eq(5, x, k_i)
    y_varI_c.append(tmp)
    plt.plot(x,tmp,label = "$K_i = $"+str(k_i))
plt.legend(ncol=3)

plt.figure(1, figsize = (10.0, 8.0))
plt.title('Michaelis-Menten Non-Competitive Kinetics with Various Inhibitor Staturation Constants')
plt.xlabel('Substrate Concentration ($\mu M$)')
plt.ylabel('Reaction Max Speed ($\mu M/s$)')

for k_i in Ki_list:
    tmp = non_competitve_kinetics_eq(5, x, k_i)
    y_varI_nc.append(tmp)
    plt.plot(x,tmp,label = "$K_i = $"+str(k_i))
plt.legend(ncol=3)

plt.figure(2, figsize = (10.0, 8.0))
plt.title('Michaelis-Menten Uncompetitive Kinetics with Various Inhibitor Staturation Constants')
plt.xlabel('Substrate Concentration ($\mu M$)')
plt.ylabel('Reaction Max Speed ($\mu M/s$)')

for k_i in Ki_list:
    tmp = un_competitve_kinetics_eq(5, x, k_i)
    y_varI_uc.append(tmp)
    plt.plot(x,tmp,label = "$K_i = $"+str(k_i))
plt.legend(ncol=3)
```

```python
k1 = 2
k2 = 2
k3 = 1
def flowfield(k1, k2, k3, steps):
    xList = []
    yList = []
    x_compList = []
    y_compList = []
    for i in range(steps):
        for j in range(steps):
            x = float(i)/(steps-1)
            y = float(j)/(steps-1)
            xList.append(x)
            yList.append(y)
            x_compList.append(-k1 * x + k3 * y)
            y_compList.append( k1 * x - (k2+k3) * y)
    plt.figure()
    plt.quiver(xList, yList, x_compList, y_compList)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
flowfield(k1, k2, k3, 21)
```
